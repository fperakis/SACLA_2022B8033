```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import h5py 
```

```python
# -- parameters
run = 347336
file_path = '/UserData/fperakis/test_data/run_%d.h5'%run # h5 files folder
bgr_path = '/UserData/fperakis/test_data/backgr_347333.h5'
```

```python
# -- import data
f = h5py.File(file_path, 'r')
list(f['data'].keys())
img = np.array(f['data/img'])

f = h5py.File(bgr_path, 'r')
list(f['data'].keys())
bgr = np.array(f['data/img'])
```

```python
scale = np.average(img[900:1150,1000:1150])/np.average(bgr[900:1150,1000:1150])
print(scale)
```

```python
plt.figure(figsize=[10,4])
plt.subplot(1,2,1)
plt.imshow(img,cmap='jet',vmin=0,vmax=380)
plt.colorbar()
plt.title('hit')

plt.subplot(1,2,2)
plt.imshow(bgr,cmap='jet',vmin=0,vmax=380)
plt.colorbar()
plt.title('miss')
plt.show()
```
